```python
plt.plot(a,h,'ro')
b1,b0=np.polyfit(a,h,deg=1) #deg=1->linear eq
print('yhat=',b0,'+',b1,'x')
yp=b0+b1*a  #yp=>y predicted(fitted value)
#plt.plot(figsize=(22,16))
plt.plot(a,yp)
plt.xlabel('age')
plt.ylabel('height')
plt.title('graph with best fit line')
plt.show()

#coefficient of determination
print(r2_score(h,yp))
rmse = mean_squared_error(h, yp)
print('rmse',rmse)
print('accuracy',r2_score(h,yp)*100)
```
